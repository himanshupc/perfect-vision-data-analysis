```python
import pandas as pd

account = pd.read_csv("account.csv")
contact = pd.read_csv("contact.csv")
lead = pd.read_csv("lead.csv")
opportunity = pd.read_csv("opportunity.csv")
order = pd.read_csv("order.csv")

# set width to all columns 
pd.set_option("display.max_columns", None)
pd.set_option("display.width", None)
pd.set_option("display.max_colwidth", None)
```

```python
# Initial shapes of all files
print("account.csv shape: ", account.shape)
print("contact.csv shape: ", contact.shape)
print("opportunity.csv shape: ", opportunity.shape)
print("lead.csv shape: ", lead.shape)
print("order.csv shape: ", order.shape)
```

```python
dfs = {
    "account": account,
    "contact": contact,
    "lead": lead,
    "opportunity": opportunity,
    "order": order
}

empty_columns = {}

for name, df in dfs.items():
    empty_columns[name] = df.columns[df.isna().all()].tolist()
```

```python
with open("empty_columns_report.txt", "w", encoding="utf-8") as f:
    for name, cols in empty_columns.items():
        f.write(f"{name}.csv\n")
        if cols:
            for col in cols:
                f.write(f"  - {col}\n")
        else:
            f.write("  (no empty columns)\n")
        f.write("\n")
```

```python
for name, df in dfs.items():
    dfs[name] = df.dropna(axis=1, how="all")
```

```python
account = dfs["account"]
contact = dfs["contact"]
lead = dfs["lead"]
opportunity = dfs["opportunity"]
order = dfs["order"]
```
